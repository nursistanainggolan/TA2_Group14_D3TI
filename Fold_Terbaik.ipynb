```python
duplicated = data_kelompok14.duplicated().sum()
if duplicated:
    print('Duplicates Row in Dataset are : {}'.format(duplicated))
else:
        print('Dataset contains no Duplicate values')
```

```python
data_kelompok14.drop_duplicates(keep='first', inplace=True)
```

```python
data_kelompok14
```

```python
x_train=data_kelompok14.values
x_train=np.delete(x_train,5,axis=1)
y_train=data_kelompok14['Target'].values
```

```python
from sklearn.naive_bayes import GaussianNB
modelnb = GaussianNB()
```

```python
x_train
```

```python
modelnb.fit(x_train,y_train)
```

```python
data_test=pd.read_excel('./data_test4.xlsx')
display(data_test)
```

```python
x_test=data_test.values
x_test=np.delete(x_test,5,axis=1)
y_test=data_test['Target'].values
```

```python
y_pred_prob=modelnb.predict_proba(x_test)
y_pred_prob
```

```python
y_test
```

```python
prediksi = modelnb.predict(x_test)
prediksi
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, prediksi)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, prediksi))
```
