```python
x=data_kelompok14.values
x=np.delete(x,5,axis=1)
y=data_kelompok14['Target'].values
```

```python
from sklearn.naive_bayes import GaussianNB
modelnb = GaussianNB()
```

```python
def get_score(modelnb,X_train, X_test, y_train, y_test):
    modelnb.fit(X_train,y_train)
    return modelnb.score(X_test,y_test)
```

```python
kf=KFold(n_splits=5, random_state=123, shuffle=True)
print(kf)
scores_nb = []
i=1
for train_index, test_index in kf.split(x,y):
    print("Fold ", i)
    print("TRAIN :", train_index, "TEST :", test_index)
    X_train=x[train_index]
    X_test=x[test_index]
    y_train=y[train_index]
    y_test=y[test_index]
    i+=1
    scores_nb.append(get_score(GaussianNB(), X_train, X_test, y_train, y_test))
    print ("Train set:",X_train.shape, y_train.shape)
    print ("Test set:",X_test.shape, y_test.shape)
```

```python
print("akurasi per fold :")
scores_nb
```
